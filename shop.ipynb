```python
from mc_optimiser.models.simple_store import SimpleShopModel

import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

# %matplotlib inline
%matplotlib widget
%load_ext autoreload
%autoreload 2
```

```python
results = []
days_to_run = 20_000
for factor in range(1, 100):
    shop = SimpleShopModel(factor/10, days_to_run)
    shop.item_cost = 50
    shop.item_price = 150
    shop.run()
    results.append(shop.revenue / days_to_run)

results = np.array(results)
results = results / np.max(results)
```

```python
plt.close()
plt.figure(figsize=(10, 5))
plt.plot(results)
plt.grid()
# plt.ylim(200, 250)
plt.xticks(range(0, 101, 5), [f'{f/10:.1f}' for f in range(0, 101, 5)])
plt.show()
```
